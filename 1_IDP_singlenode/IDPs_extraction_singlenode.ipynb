```python
import dxpy
import subprocess

# 自动发现已分发数据集ID并加载数据集
dispensed_dataset_id = dxpy.find_one_data_object(typename='Dataset', name='app*.dataset', folder='/', name_mode='glob')['id']

# 获取项目ID
project_id = dxpy.find_one_project()["id"]
dataset = (':').join([project_id, dispensed_dataset_id])

cmd = ["dx", "extract_dataset", dataset, "-ddd", "--delimiter", ","]
subprocess.check_call(cmd)
```

```python
import os
import glob
import pandas as pd

path = os.getcwd()

# 查找当前目录下的 .data_dictionary.csv 文件
data_dict_csv = glob.glob(os.path.join(path, "*.data_dictionary.csv"))[0]

data_dict_df = pd.read_csv(data_dict_csv, low_memory=False)
data_dict_df.head()
```

```python
import pandas as pd
import subprocess
import time

start_index = 19800
step = 1100  # 每次切片的步长
iterations = 10  # 需要运行的次数，可调整

# 获取字段名
field_names = list(
    data_dict_df.loc[data_dict_df['folder_path'].str.contains('Assessment centre > Imaging > Brain MRI', na=False), "name"].values
)



for i in range(iterations):
    # 计算当前批次的索引范围
    end_index = start_index + step - 1
    
    # 生成格式化后的字段名称字符串
    field_names_str = [f"participant.{f}" for f in field_names]
    field_names_IDPs = ",".join(field_names_str)
    field_names_IDPs = field_names_IDPs[start_index:end_index]
    print(field_names_IDPs)
    
    # 处理数据并转换为 DataFrame
    field_names_list = field_names_IDPs.split(',')
    field_names_df = pd.DataFrame(field_names_list, columns=['Field Names'])
    field_names_df = field_names_df[:30]  # 取前 30 行
    print(field_names_df)

    # 保存当前批次的字段名到 CSV
    field_names_csv = f'field_names_IDPs_{i}.csv'
    field_names_df.to_csv(field_names_csv, index=False)
    
    # 运行 dx 命令，提取数据集
    cmd = [
        'dx',
        'extract_dataset',
        dataset,
        '--fields',
        field_names_IDPs,
        '--delimiter',
        ',',
        '--output',
        f'extracted_data_{i}.csv',  # 使用 f-string 格式化文件名
    ]
    
    start_time = time.time()
    subprocess.check_call(cmd)  # 运行命令
    execution_time = time.time() - start_time
    print(f"dx 命令执行的运行时间 (批次 {i}): {execution_time:.2f} 秒")
    
    # 更新索引
    start_index = end_index + 1
```
